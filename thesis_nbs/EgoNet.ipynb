```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

```python
def draw_ego(G, i, radius=2, **kwargs):
    ego = nx.ego_graph(G, i, radius=radius)
    pos = nx.spring_layout(ego, seed=1337)
    nx.draw(ego, pos, node_color=['black' if i != v else 'red' for v in ego.nodes], **kwargs)
```

```python
G = nx.generators.community.stochastic_block_model(
    [15, 5, 10, 15],
    [[.3, .01, .02, .01],
     [.01, .03, .01, .1],
     [.02, .01, .4, .0],
     [.01, .1, .0, .2],], seed=1337,
)
pos = nx.spring_layout(G, seed=1337)
np.random.seed(1337)
pos = {
    k : v + 0.03 * np.random.randn(2) for k, v in pos.items()
}
```

```python
pos_arr = np.array([p for p in pos.values()])
x_min, x_max = pos_arr[:, 0].min(), pos_arr[:, 0].max()
y_min, y_max = pos_arr[:, 1].min(), pos_arr[:, 1].max()

x_min -= 0.1 * (x_max - x_min)
x_max += 0.1 * (x_max - x_min)
y_min -= 0.1 * (y_max - y_min)
y_max += 0.1 * (y_max - y_min)
```

```python
fig, ax = plt.subplots()
nx.draw(G, pos, node_color='black', width=.5, node_size=50, ax=ax)
ax.set_xlim([x_min, x_max])
ax.set_ylim([y_min, y_max])

ax.set_axis_on()
fig.savefig('egonet_full.pdf', bbox_inches='tight')
```

```python
idx = 25
ego = nx.ego_graph(G, idx, radius=2)
fig, ax = plt.subplots()


nx.draw(G, pos, node_color='lightgrey', width=.5, node_size=50, ax=ax,
           nodelist = [v for v in G.nodes if v not in ego.nodes],
            edgelist = [e for e in G.edges if e not in ego.edges],
            edge_color='lightgrey',
       )

nx.draw(G, pos, node_color=['black' if v != idx else 'red' for v in ego.nodes], 
        width=.5, nodelist=ego.nodes, edgelist=ego.edges, ax=ax, node_size=50)

ax.set_xlim([x_min, x_max])
ax.set_ylim([y_min, y_max])
ax.set_axis_on()
fig.savefig('egonet_1.pdf', bbox_inches='tight')
```
